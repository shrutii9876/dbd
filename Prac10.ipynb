```python
features=iris.columns[:-1]
feature_types=iris[features].dtypes
```

```python
outliers={}
for feature in features:
    Q1=iris[feature].quantile(0.25)
    Q3=iris[feature].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    outlier_list=iris[(iris[feature]<lower_bound) | (iris[feature]>upper_bound)]
    outliers[feature] = outlier_list.copy() 
```

```python
print("Outliers:")
for feature,outlier in outliers.items():
    print(f"{feature}:{outlier.index.tolist()}")
```
